```python
import tkinter as tk
import requests
from PIL import Image, ImageTk
from io import BytesIO

def fetch_image(url):
    response = requests.get(url)
    image = Image.open(BytesIO(response.content))
    print(image)
    return image

def display_image(image):
    window = tk.Tk()
    window.title("AI成图")
    window.geometry("600x400")

    img = ImageTk.PhotoImage(image)
    label = tk.Label(window, image=img)
    label.pack()

    window.mainloop()

# 测试示例
image = fetch_image(url)
display_image(image)
```

```python
import tkinter as tk
import os
import openai
import json
import sys
import requests
from PIL import ImageTk, Image
import time
import webbrowser
import urllib

def download_image(url, save_path):
    urllib.request.urlretrieve(url, save_path)

def request_picture():
    global new_image
    global user_input 
    user_input = entry.get()
    # Load your API key from an environment variable or secret management service
    openai.api_key = "#"
    response = openai.Image.create(prompt=user_input,size="1024x1024")
    url = response["data"][0]["url"]
    return url

def handle_button_click():
    url = request_picture()
    input_save_path = user_input+".jpg"
    download_image(url, input_save_path)
    webbrowser.open(url)
    
window = tk.Tk()
window.configure(bg="blue")
window.title("AI作画")
window.geometry("600x500")
window.iconbitmap("yes.ico")  # 设置小图标
# 添加背景图

image_path = "batman.png"  # 背景图路径
image = Image.open(image_path)
background_image = ImageTk.PhotoImage(image)
background_label = tk.Label(window, image=background_image)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

label = tk.Label(window, text="请描述画面内容:",font=("Arial", 20, "bold"))
label.pack()

entry = tk.Entry(window, width=60)
entry.pack()

button = tk.Button(window, text="提交", command=handle_button_click,font=("Arial", 20, "bold"))
button.pack()

window.mainloop()
```
