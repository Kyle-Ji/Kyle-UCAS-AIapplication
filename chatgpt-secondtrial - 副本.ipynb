```python
import tkinter as tk
from tkinter import messagebox
import os
import openai
from PIL import Image, ImageTk

openai.api_key = "#"

def send_message():
    user_message = user_input.get()  
    # 加载图片并调整大小
    image = Image.open("user.png")
    image = image.resize((30, 30))
    # 创建图片标签
    photo = ImageTk.PhotoImage(image)
    label = tk.Label(image=photo)
    label.image = photo  # 需要保留对图片的引用，否则显示不出来
    chat_log.window_create(tk.END, window=label)
    chat_log.insert(tk.END, "User: " + user_message + "\n\n")
    # 调用机器人回复函数，得到机器人的回复内容
    robot_reply = get_reply(user_message)
     # 加载图片并调整大小
    image1 = Image.open("AI.png")
    image1 = image1.resize((30, 30))
    # 创建图片标签
    photo1 = ImageTk.PhotoImage(image1)
    label1 = tk.Label(image=photo1)
    label1.image = photo1  # 需要保留对图片的引用，否则显示不出来
    chat_log.window_create(tk.END, window=label1)
    chat_log.insert(tk.END, "Robot: " + robot_reply + "\n\n")
    # 清空用户输入框的内容
    user_input.delete(0, tk.END)
    
def get_reply(message):
    # 在这里编写机器人回复的逻辑
    completion = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": message},
    {"role": "user", "content": ""}
  ]
)
    return completion.choices[0].message["content"]
    
# 创建主窗口
window = tk.Tk()
window.title("chatbot")

window.iconbitmap("yes.ico")  # 设置小图标

# 创建对话展示区域
chat_log = tk.Text(window, width=100, height=40)
chat_log.pack()

# 创建底部区域容器
bottom_frame = tk.Frame(window)
bottom_frame.pack(pady=10)

# 创建用户输入框
user_input = tk.Entry(bottom_frame, width=40, font=("Arial", 20))
user_input.pack(side=tk.LEFT, padx=5)

# 创建发送按钮
send_button = tk.Button(bottom_frame, text="发送", font=("Arial", 20), command=send_message)
send_button.pack(side=tk.LEFT, padx=5)

# 设置输入框和发送按钮的样式
user_input.config(relief=tk.SOLID, borderwidth=1, font=("Arial", 20), bg="white", fg="black")
send_button.config(relief=tk.SOLID, borderwidth=1, font=("Arial", 20), bg="#4CAF50", fg="white", bd=0, padx=10, pady=5, cursor="hand2")

# 将输入框和发送按钮设计为圆角矩形
user_input.config(highlightthickness=0, highlightbackground="gray", highlightcolor="gray", bd=0)
send_button.config(highlightthickness=0, bd=0, highlightbackground="#4CAF50", highlightcolor="#4CAF50")

# 运行程序
window.mainloop()
```
